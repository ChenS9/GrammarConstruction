##### get prepositional/Existential/adverb words

```python
def get_existential(text):
    blob = TextBlob(text)
    return [ word for (word,tag) in blob.tags if tag == "EX"]

df_text['EX'] = df_text['final'].apply(get_existential)
```

```python
def get_adverb(text):
    blob = TextBlob(text)
    return [ word for (word,tag) in blob.tags if tag == "RB"]

df_text['RB'] = df_text['final'].apply(get_adverb)
```

```python
def get_preposition(text):
    blob = TextBlob(text)
    return [ word for (word,tag) in blob.tags if tag == "IN"]

df_text['IN'] = df_text['final'].apply(get_preposition)
```

### Method 2: nltk collocations
#### Trigram with nltk collocations

```python
def nbest(self, score_fn, n):
    """Returns the top n ngrams when scored by the given function."""
    return [p for p,s in self.score_ngrams(score_fn)[:n]]
```

```python
def score_ngrams(self, score_fn):
    """Returns a sequence of (ngram, score) pairs ordered from highest to
    lowest score, as determined by the scoring function provided.
    """
    return sorted(self._score_ngrams(score_fn),
                  key=_itemgetter(1), reverse=True)
```

```python
trigram_measures = nltk.collocations.TrigramAssocMeasures()
```
